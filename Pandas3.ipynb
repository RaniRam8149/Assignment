```python
Consider following code to answer further questions:
import pandas as pd
course_name = [‘Data Science’, ‘Machine Learning’, ‘Big Data’, ‘Data Engineer’]
duration = [2,3,6,4]
df = pd.DataFrame(data = {‘course_name’ : course_name, ‘duration’ : duration})
Q1. Write a code to print the data present in the second row of the dataframe, df.
import pandas as pd

course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']
duration = [2, 3, 6, 4]

df = pd.DataFrame(data={'course_name': course_name, 'duration': duration})

print(df.iloc[1])

Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?

In pandas, loc and iloc are both used for accessing data within a DataFrame, but they differ in how they reference the data:

loc: This is primarily label-based indexing. It is used to access rows and columns by their labels or boolean array.

iloc: This is primarily integer-based indexing. It is used to access rows and columns by their integer position, similar to Python's standard list slicing.

Here's a breakdown of their differences:

loc:

Accesses data based on label.
Uses the row and column labels to access data.
The row labels are typically index labels, and the column labels are column names.
Allows slicing using label names.
Inclusive of both start and stop labels when using slicing.
Raises KeyError if the specified labels are not found.
iloc:

Accesses data based on integer position.
Uses integer positions to access data.
The integer positions start from 0.
Allows slicing using integer positions.
Exclusive of the stop position when using slicing (similar to Python slicing convention).
Raises IndexError if the specified integer positions are out of range.

Q3. Reindex the given dataframe using a variable, reindex = [3,0,1,2] and store it in the variable, new_df
then find the output for both new_df.loc[2] and new_df.iloc[2].
import pandas as pd

course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']
duration = [2, 3, 6, 4]
df = pd.DataFrame(data={'course_name': course_name, 'duration': duration})

reindex = [3, 0, 1, 2]
new_df = df.reindex(reindex)

print("Output for new_df.loc[2]:")
print(new_df.loc[2])

print("\nOutput for new_df.iloc[2]:")
print(new_df.iloc[2])

Q4. Write a code to find the following statistical measurements for the above dataframe df1:
(i) mean of each and every column present in the dataframe.
(ii) standard deviation of column, ‘column_2’
import pandas as pd
import numpy as np

columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
indices = [1, 2, 3, 4, 5, 6]
df1 = pd.DataFrame(np.random.rand(6, 6), columns=columns, index=indices)

column_means = df1.mean()
print("Mean of each column:")
print(column_means)

std_dev_column_2 = df1['column_2'].std()
print("\nStandard deviation of column 'column_2':", std_dev_column_2)

Explanation:

Mean of each column (column_means):

Using the .mean() method on the DataFrame calculates the mean for each column.
The result is a Series where the index represents the column names and the values represent the mean of each column.
Standard deviation of column 'column_2' (std_dev_column_2):

Using the .std() method on a specific column, in this case, 'column_2', calculates the standard deviation for that column.
The result is a single value representing the standard deviation of 'column_2'.

Q5. Replace the data present in the second row of column, ‘column_2’ by a string variable then find the
mean of column, column_2.
If you are getting errors in executing it then explain why.
[Hint: To replace the data use df1.loc[] and equate this to string data of your choice.]
import pandas as pd
import numpy as np

columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
indices = [1, 2, 3, 4, 5, 6]
df1 = pd.DataFrame(np.random.rand(6, 6), columns=columns, index=indices)

df1.loc[2, 'column_2'] = 'StringData'

mean_column_2 = df1['column_2'].mean()
print("Mean of 'column_2':", mean_column_2)

Executing this code will result in an error like: TypeError: unsupported operand type(s) for +=: 'float' and 'str'. This error occurs because you're trying to add a string ('StringData') to a float value during the mean calculation, which is not allowed in Python.

Q6. What do you understand about the windows function in pandas and list the types of windows
functions?
In pandas, the window functions provide a flexible way to perform rolling or expanding calculations on data within a specified window or range. These functions are particularly useful for time series data analysis and can also be applied to other types of data.

Window functions typically involve calculating some statistic (e.g., mean, sum, standard deviation) over a specified window of data points. The window can be defined based on a fixed number of preceding or following data points, or it can be based on a fixed time period.

Types of window functions in pandas include:

Rolling Window Functions:

These functions perform calculations over a fixed-size window of data points.
Commonly used rolling window functions include rolling.mean(), rolling.sum(), rolling.std(), etc.
Example: Calculating the rolling average of a time series over the past 7 days.
Expanding Window Functions:

These functions perform calculations over an expanding window that grows with the data.
The expanding window includes all data points up to the current point.
Commonly used expanding window functions include expanding.mean(), expanding.sum(), expanding.std(), etc.
Example: Calculating the cumulative sum of a time series over time.
Exponentially Weighted Moving Average (EWMA):

EWMA is a type of rolling window function that assigns exponentially decreasing weights to past data points.
It gives more weight to recent data points and less weight to older data points.
Implemented using the ewm() function in pandas.
Example: Calculating the EWMA of a time series to smooth out fluctuations.
Window Aggregations:

These functions aggregate data over a fixed-size or variable-size window.
They allow custom aggregation functions to be applied over a window of data.
Implemented using the rolling.apply() or expanding.apply() functions with a custom aggregation function.
Example: Calculating the median of a time series over a rolling window.

Q7. Write a code to print only the current month and year at the time of answering this question.
[Hint: Use pandas.datetime function]
import pandas as pd
from datetime import datetime

current_date = datetime.now()

current_month = current_date.month
current_year = current_date.year

print("Current month:", current_month)
print("Current year:", current_year)

Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and
calculates the difference between them in days, hours, and minutes using Pandas time delta. The
program should prompt the user to enter the dates and display the result.
import pandas as pd

def calculate_time_difference(start_date, end_date):
   
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)
    
    time_diff = end_date - start_date
    
    days_diff = time_diff.days
    hours_diff = time_diff.seconds // 3600
    minutes_diff = (time_diff.seconds % 3600) // 60
    
    return days_diff, hours_diff, minutes_diff

start_date = input("Enter the start date (YYYY-MM-DD): ")
end_date = input("Enter the end date (YYYY-MM-DD): ")

days_diff, hours_diff, minutes_diff = calculate_time_difference(start_date, end_date)

print("Time difference:")
print("Days:", days_diff)
print("Hours:", hours_diff)
print("Minutes:", minutes_diff)

Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified
column to a categorical data type. The program should prompt the user to enter the file path, column
name, and category order, and then
import pandas as pd

def convert_to_categorical(df, column_name, category_order):
    
    df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)
    return df

file_path = input("Enter the file path of the CSV file: ")

try:
    df = pd.read_csv(file_path)
except FileNotFoundError:
    print("File not found. Please enter a valid file path.")
    exit()

print("Columns in the DataFrame:")
print(df.columns)

column_name = input("Enter the name of the column you want to convert to categorical: ")

if column_name not in df.columns:
    print("Column '{}' not found in the DataFrame.".format(column_name))
    exit()

category_order = input("Enter the category order (comma-separated): ").split(',')

df = convert_to_categorical(df, column_name, category_order)

print("\nUpdated DataFrame:")
print(df)

Q10. Write a Python program that reads a CSV file containing sales data for different products and
visualizes the data using a stacked bar chart to show the sales of each product category over time. The
program should prompt the user to enter the file path and display the chart.
import pandas as pd
import matplotlib.pyplot as plt

def visualize_sales_data(file_path):
   
    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        print("File not found. Please enter a valid file path.")
        return
    
    if 'Date' in df.columns:
        df['Date'] = pd.to_datetime(df['Date'])

    df_pivot = df.pivot_table(index='Date', columns='Product Category', values='Sales', aggfunc='sum', fill_value=0)

    df_pivot.plot(kind='bar', stacked=True, figsize=(10, 6))
    plt.title('Sales of Each Product Category Over Time')
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.xticks(rotation=45)
    plt.legend(title='Product Category', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

file_path = input("Enter the file path of the CSV file containing sales data: ")

visualize_sales_data(file_path)

Q11. You are given a CSV file containing student data that includes the student ID and their test score. Write
a Python program that reads the CSV file, calculates the mean, median, and mode of the test scores, and
displays the results in a table.
The program should do the followingM
I Prompt the user to enter the file path of the CSV file containing the student dataR
I Read the CSV file into a Pandas DataFrameR
I Calculate the mean, median, and mode of the test scores using Pandas toolsR
I Display the mean, median, and mode in a table.
Assume the CSV file contains the following columnsM
I Student ID: The ID of the studentR
I Test Score: The score of the student's test.
Example usage of the program:
Enter the file path of the CSV file containing the student data: student_data.csv
+-----------+--------+
| Statistic | Value |
+-----------+--------+
| Mean | 79.6 |
| Median | 82 |
| Mode | 85, 90 |
+-----------+--------+
Assume that the CSV file student_data.csv contains the following data:
Student ID,Test Score
1,85
2,90
3,80
4,75
5,85
6,82
7,78
8,85
9,90
10,85
The program should calculate the mean, median, and mode of the test scores and display the results
in a table.
import pandas as pd

def calculate_statistics(file_path):
    
    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        print("File not found. Please enter a valid file path.")
        return

    mean_score = df['Test Score'].mean()
    median_score = df['Test Score'].median()
    mode_scores = df['Test Score'].mode()

    results = pd.DataFrame({
        'Statistic': ['Mean', 'Median', 'Mode'],
        'Value': [mean_score, median_score, ', '.join(map(str, mode_scores))]
    })

    print(results.to_string(index=False))


if __name__ == "__main__":
    file_path = input("Enter the file path of the CSV file containing the student data: ")
    calculate_statistics(file_path)


```
